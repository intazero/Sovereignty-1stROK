## 조선일보 데이터

```python
df_chosun = pd.read_csv("C:/Users/Admin/Downloads/preprocess/chosun_preprocessed.csv", encoding = 'utf-8')
df_chosun
```

```python
df_chosun.info()
```

```python
# 필요없는 column 모두 제거.
# 기사 제목, 제목 및 본문의 경우 'title_archaic', 'body_archaic' 만 남기고 모두 삭제 (_korean의 경우 _archaic에 비해 NaN이 더 많아 기준으로 두기 어려움)
df_chosun_modified = df_chosun.drop(['Unnamed: 0', 'id', 'page_no', 'regdate', 'type', 'publication_day', 'mes', 'section',
                                     'completeness', 'news_name', 'publication_no', 'node_id', 'source_image_file', '@timestamp',
                                     'print_no', '@version', 'outline', 'isn', 'isn_update', 'page_section', 'bfly_newline', 'url',
                                     'image', 'site', 'yonsei_merge', 'source_xml_file', 'hanja_trans_type', 'path',
                                     'host', 'title_archaic_hangul', 'title_hanja_korean', 'title_korean',
                                     'subtitle_hanja_korean', 'subtitle_korean', 'subtitle_archaic_hangul',
                                     'subtitle_archaic', 'body_archaic_hangul', 'body_hanja_korean', 'body_korean',
                                    'authors_name', 'authors_name_hanja'], axis = 1)
df_chosun_modified
```

```python
# 제1공화국 시기(1948.08.15.~1960.06.14.) 기사를 필터링
df_chosun_thesis = df_chosun_modified[df_chosun_modified['publication_date'].between('1948-08-15', '1960-06-14')]
df_chosun_thesis
```

```python
df_chosun_thesis_revised = df_chosun_thesis[df_chosun_thesis.body_archaic.isna() != True] # 'body_archaic'이 NaN인 기사 제거
df_chosun_thesis_revised = df_chosun_thesis_revised.rename(columns = {'title_archaic' : 'title_ori'}) # 동아일보 dataframe에 맞추어 column 이름 편집
df_chosun_thesis_revised = df_chosun_thesis_revised.rename(columns = {'body_archaic' : 'body_ori'})
df_chosun_thesis_revised = df_chosun_thesis_revised[['title_ori', 'body_ori', 'publication_date', 'year']]
df_chosun_thesis_revised['company'] = '조선일보' # 'company' 열 추가
df_chosun_thesis_revised
```

## 동아일보 데이터

```python
df_donga = pd.read_csv("C:/Users/Admin/Downloads/preprocess/donga_preprocessed.csv", encoding = 'utf-8')
df_donga
```

```python
df_donga.info()
```

```python
# 제1공화국 시기 기사를 필터링
df_donga_thesis = df_donga[df_donga['publication_date'].between('1948-08-15', '1960-06-14')]
df_donga_thesis
```

```python
# 필요없는 column 모두 제거. 제목과 본문의 경우 'title_ori', 'body_ori'만 남기고 모두 제거
df_donga_thesis_revised = df_donga_thesis.drop(['Unnamed: 0', 'idx', 'title_trans', 'page', 'reporter', 'body_trans'], axis = 1)
df_donga_thesis_revised['company'] = '동아일보' # 'company' 열 추가
df_donga_thesis_revised
```

## 동아일보, 조선일보 데이터 병합 및 저장

```python
df_chosun_thesis_revised
```

```python
df_donga_thesis_revised
```

```python
# 동아일보, 조선일보 데이터 병합
result = pd.concat([df_chosun_thesis_revised, df_donga_thesis_revised])
result
```
